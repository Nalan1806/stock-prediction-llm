```python
import yfinance as yf
import pandas as pd

# Download historical data for AAPL
data = yf.download("AAPL", start="2020-01-01", end="2025-01-01")

# Display the first few rows and the columns to understand the structure
print("Data Head:")
display(data.head())
print("\nData Columns:")
print(data.columns)

# Summarize key statistics for relevant columns
print("\nBasic Statistics:")
# Access columns using the correct MultiIndex tuples
print(data[[('Open', 'AAPL'), ('High', 'AAPL'), ('Low', 'AAPL'), ('Close', 'AAPL'), ('Volume', 'AAPL')]].agg(['mean', 'median', 'min', 'max', 'std']))
```

```python
# Calculate Technical Indicators

# Relative Strength Index (RSI)
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).fillna(0)
    loss = (-delta.where(delta < 0, 0)).fillna(0)
    avg_gain = gain.ewm(com=window-1, adjust=False).mean()
    avg_loss = loss.ewm(com=window-1, adjust=False).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

data['RSI'] = calculate_rsi(data)

# Moving Average Convergence Divergence (MACD)
exp1 = data['Close'].ewm(span=12, adjust=False).mean()
exp2 = data['Close'].ewm(span=26, adjust=False).mean()
data['MACD'] = exp1 - exp2
data['Signal Line'] = data['MACD'].ewm(span=9, adjust=False).mean()

# Bollinger Bands
window = 20
data['Rolling Mean'] = data['Close'].rolling(window=window).mean()
data['Rolling Std'] = data['Close'].rolling(window=window).std()
data['Upper Band'] = data['Rolling Mean'] + (data['Rolling Std'] * 2)
data['Lower Band'] = data['Rolling Mean'] - (data['Rolling Std'] * 2)


# Calculate Volatility Measures

# Average True Range (ATR)
def calculate_atr(data, window=14):
    high_low = data['High'] - data['Low']
    high_close = abs(data['High'] - data['Close'].shift())
    low_close = abs(data['Low'] - data['Close'].shift())
    ranges = pd.concat([high_low, high_close, low_close], axis=1)
    true_range = ranges.max(axis=1)
    atr = true_range.ewm(com=window-1, adjust=False).mean()
    return atr

data['ATR'] = calculate_atr(data)

# Display the data with the new feature columns
print("Data with Technical Indicators and Volatility Measures:")
display(data.tail())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Prepare data for Regression
# Combine features and target to drop NaNs consistently
data_reg = data.copy()
data_reg['Target_Close'] = data_reg['Close'] # Use a temporary column for the target

# Features are all columns except the temporary target
X = data_reg.drop(['Close', 'Target_Close'], axis=1)
y = data_reg['Target_Close']

# Drop NaNs from the combined data
combined_data_reg = pd.concat([X, y], axis=1).dropna()
X = combined_data_reg.drop('Target_Close', axis=1)
y = combined_data_reg['Target_Close']

# Split data for Regression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Regression Data Shapes:")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

# Train Linear Regression Model
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, y_train)

# Prepare data for Classification
# Create binary target variable: 1 if next day's closing price is higher, 0 other
data_cls = data.copy()
data_cls['Price_Change'] = data_cls['Close'].pct_change().shift(-1)
data_cls['Target_Classification'] = np.where(data_cls['Price_Change'] > 0, 1, 0) # Remove dropna() here

# Features (X_cls) are all columns except the targets and future price change
X_cls = data_cls.drop(['Close', 'Price_Change', 'Target_Classification'], axis=1)
y_cls = data_cls['Target_Classification']

# Drop NaNs from the combined data for classification
combined_data_cls = pd.concat([X_cls, y_cls], axis=1).dropna()
X_cls = combined_data_cls.drop('Target_Classification', axis=1)
y_cls = combined_data_cls['Target_Classification']


# Split data for Classification
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_cls, y_cls, test_size=0.3, random_state=42)

print("\nClassification Data Shapes:")
print("X_train_cls shape:", X_train_cls.shape)
print("X_test_cls shape:", X_test_cls.shape)
print("y_train_cls shape:", y_train_cls.shape)
print("y_test_cls shape:", y_test_cls.shape)


# Train Random Forest Classifier Model
random_forest_model = RandomForestClassifier(random_state=42)
random_forest_model.fit(X_train_cls, y_train_cls)
```
